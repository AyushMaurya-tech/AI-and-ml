```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img(r"E:\ML and AI\deep learning\neural network\cnn\cat\cat\cat.webp",target_size=(200,200))
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    fill_mode = 'nearest'

)
```

```python
img = image.img_to_array(img)
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i = 0 
for output in datagen.flow(input_batch,batch_size=1,save_to_dir=r"E:\ML and AI\deep learning\neural network\cnn\cat\arg cat"):
    i = i+1
    if i == 10:
        break
```
