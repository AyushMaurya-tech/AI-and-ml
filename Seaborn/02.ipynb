```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
tips_df = sns.load_dataset('tips')
crash_df = sns.load_dataset('car_crashes')
sns.pairplot(tips_df,hue = 'sex',palette='Blues')
```

Rug plot

```python
sns.rugplot(tips_df['tip'])
```

Styling

```python
sns.set_style('ticks')
plt.figure(figsize=(8,4))
sns.set_context('paper',font_scale = 2)
sns.jointplot(x = 'speeding',y = 'alcohol',data = crash_df,kind='reg')
#sns.despine(left=True,bottom=True)#axis gayab
```

Categorical plots

Bar Plot

```python
sns.barplot(x =tips_df.sex,y = tips_df.total_bill,estimator=np.median)
```

count plot

```python
sns.countplot(x = tips_df.sex)
```

Box plot

```python
sns.boxplot(x = tips_df.day, y = tips_df.total_bill,hue = tips_df.sex)
plt.legend(loc = 0)
```

violin plot

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips_df,hue = 'sex',split = True)
```

strip plot

```python
plt.figure(figsize= (8,5))
sns.stripplot(x = tips_df.day, y = tips_df.total_bill,jitter=True,hue= tips_df.sex,dodge=True) 
```
