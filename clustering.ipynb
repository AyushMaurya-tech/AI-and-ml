```python
numeric_columns = iris_df.select_dtypes(include=['number']).columns
numeric_columns_list = numeric_columns.tolist()
numeric_columns_list
```

K_means clustering

```python
from sklearn.cluster import KMeans
```

```python
model1 = KMeans(random_state=42,n_clusters=3)
```

```python
iris_df[numeric_columns]
```

```python
model1.fit(iris_df[numeric_columns])
```

```python
model1.cluster_centers_
```

```python
preds = model1.predict(iris_df[numeric_columns])
preds
```

```python
sns.scatterplot(data=iris_df[numeric_columns],x = 'sepal_length',y = 'petal_length',hue=preds)
centeres_x,centeres_y = model1.cluster_centers_[:,0],model1.cluster_centers_[:,2]
plt.plot(centeres_x,centeres_y,'xb')
```

```python
model1.inertia_
```

```python
model = KMeans(n_clusters=6,random_state=42).fit(iris_df[numeric_columns])
```

```python
preds = model.predict(iris_df[numeric_columns])
preds
```

```python
sns.scatterplot(data=iris_df[numeric_columns],x = 'sepal_length',y = 'petal_length',hue=preds)
```

```python
model.inertia_
```

```python
option = range(2,11)
intertia = []
for n_clusters in option:
    model = KMeans(n_clusters,random_state=43).fit(iris_df[numeric_columns])
    intertia.append(model.inertia_)
plt.title('NO of cluster vs. INertia')
plt.plot(option,intertia,'-o')
plt.xlabel('No of clusters (K)')
plt.ylabel('inertia')
```

DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN(eps = 0.6,min_samples=15).fit(iris_df[numeric_columns])
```

```python
model.core_sample_indices_
```

```python
model.labels_
```

```python
sns.scatterplot(data=iris_df[numeric_columns],x = 'sepal_length',y = 'petal_length',hue=model.labels_)
```
