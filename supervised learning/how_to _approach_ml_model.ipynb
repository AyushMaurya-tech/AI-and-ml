```python
input_cols = ['Store', 'DayOfWeek', 'Promo', 'StateHoliday', 'StoreType', 'Assortment', 'Day', 'Month', 'Year']
```

```python
target_col = 'Sales'
```

```python
val_inputs = val_df[input_cols].copy()
val_targets = val_df[target_col].copy()
```

```python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_col].copy()
```

```python
test_inputs = merged_test_df[input_cols].copy()
# Test data does not have targets
```

```python
numeric_cols = ['Store', 'Day', 'Month', 'Year']
categorical_cols = ['DayOfWeek', 'Promo', 'StateHoliday', 'StoreType', 'Assortment']
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean').fit(train_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols] = imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = imputer.transform(test_inputs[numeric_cols])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler().fit(train_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder
categorical_Cols = categorical_cols
```

```python
train_inputs
```

```python
# Ensure all categorical columns are strings
train_inputs[categorical_cols] = train_inputs[categorical_cols].astype(str)
val_inputs[categorical_cols] = val_inputs[categorical_cols].astype(str)
test_inputs[categorical_cols] = test_inputs[categorical_cols].astype(str)

# Handle missing values
train_inputs[categorical_cols] = train_inputs[categorical_cols].fillna("Missing")
val_inputs[categorical_cols] = val_inputs[categorical_cols].fillna("Missing")
test_inputs[categorical_cols] = test_inputs[categorical_cols].fillna("Missing")

# Fit and transform using OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore').fit(train_inputs[categorical_cols])
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))

train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

```python
X_train = train_inputs[numeric_cols + encoded_cols]
X_val = val_inputs[numeric_cols + encoded_cols]
X_test = test_inputs[numeric_cols + encoded_cols]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
```

```python
val_preds = linreg.predict(X_val)
error_val_square = mean_squared_error(val_preds,val_targets)
error_val  = error_val_square**0.5
error_val
```

```python
def try_model(model):
    model.fit(X_train,train_targets)
    train_preds = model.predict(X_train)
    val_preds = model.predict(X_val)
    x = mean_squared_error(train_targets,train_preds)
    train_rmse = x**0.5
    y = mean_squared_error(val_targets,val_preds)
    val_rmse = y**0.5
    return train_rmse,val_rmse
```

REGULIZATION AND ENSEMBLING

k fold cross validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

# Load the dataset
X = X_train
y = train_targets

# Initialize the classifier
model = RandomForestRegressor(n_estimators=20, random_state=42, n_jobs=-1)


# Define the number of folds
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Perform cross-validation
scores = cross_val_score(model, X, y, cv=kf)

# Print the scores
print(f"Cross-validation scores: {scores}")
print(f"Average accuracy: {scores.mean()}")
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import make_scorer

# Load the dataset
X = X_train
y = train_targets

# Initialize the regressor
model = RandomForestRegressor(n_estimators=20, random_state=42, n_jobs=-1)

# Define the number of folds
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Define a custom scorer for Mean Squared Error (MSE)
mse_scorer = make_scorer(mean_squared_error)

# Perform cross-validation using MSE as the scoring metric
mse_scores = cross_val_score(model, X, y, cv=kf, scoring=mse_scorer)

# Print the MSE scores
print(f"Cross-validation MSE scores: {mse_scores}")
Z = mse_scores.mean()
print(Z**0.5)
```
