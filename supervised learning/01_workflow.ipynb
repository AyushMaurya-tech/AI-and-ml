VALUE IMPUTATION BY SIMPLE IMPUTER

```python
from sklearn.impute import SimpleImputer
```

```python
print(f'number of null value before imputing:{df.age.isnull().sum()}')
```

```python
imp = SimpleImputer(strategy='mean')
df['age'] = imp.fit_transform(df[['age']])
print('number of null value after imputing ')
print(df.age.isnull().sum())
```

```python
def get_parameters(df):
    parameters = {}
    for col in df.columns[df.isnull().any()]:
        if df[col].dtype =='float64' or df[col].dtype == 'int64' or df[col].dtype == 'int 32':
            strategy = 'mean'
        else:
            strategy = 'most_frequent'
        missing_values = df[col][df[col].isnull()].values[0]
        parameters[col] = {'missing_values':missing_values,'strategy':  strategy} 
    return parameters 

    

```

```python
parameters = get_parameters(df)
parameters
```

```python
for col,param in parameters.items():
    missing_values = param['missing_values']
    strategy = param['strategy']
    imp = SimpleImputer(missing_values=missing_values,strategy=strategy)
    df[col] = imp.fit_transform(df[[col]]).ravel()
```

```python
df.isnull().sum()
```

WE MERGE SIBSP AND PARCH VALUE TOGETHER

```python
df['family'] = df['sibsp'] + df['parch']
df.loc[df['family']>0,'travelled alone'] = 0
df.loc[df['family']==0,'travelled alone'] = 1
df['travelled alone'].value_counts().plot(title = 'passenger travelled alone?',kind = 'bar')
```
