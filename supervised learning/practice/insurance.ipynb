```python
import numpy as np
```

```python
def rmse(targets,prediction):
    return np.sqrt(np.mean(np.square(targets-prediction)))
```

```python
def estimate_chargesxc(age, w, b):
    return w * age + b
```

```python
targets = non_smoker_df['charges']
age = non_smoker_df['age']
predicted = estimate_chargesxc(age,w,b)
```

```python
rmse(targets,predicted)
```

```python
def try_parameters2(w, b):
    
    age = non_smoker_df['age']
    target = non_smoker_df['charges']
    predictions = estimate_chargesxc(age,w,b)

    
    # Plotting the estimated charges as a red line
    plt.plot(age, predictions, 'r', alpha=0.9, label='Estimate')
    
    # Scatter plot of the actual charges
    plt.scatter(age, target, s=8, alpha=0.9, label='Actual')
    
    # Adding labels and a legend
    plt.xlabel('Age')
    plt.ylabel('Charges')
    plt.legend()
    loss = rmse(target,predictions)
    # Displaying the plot
    plt.show()
    print(loss)

# Calling the function with w=60 and b=200
try_parameters2(60, 200)
```
