```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import keras
import tensorflow
```

```python
yxdf = pd.read_csv('Churn_Modelling.csv')
```

```python
yxdf
```

```python
xdf = pd.get_dummies(yxdf,columns=['Geography','Gender'],drop_first=True)
```

```python
xdf
```

```python
xdf[['Geography_Germany', 'Geography_Spain', 'Gender_Male']] = xdf[['Geography_Germany', 'Geography_Spain','Gender_Male']].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = xdf.drop(columns=['Surname'])
df = xdf.drop(columns=['RowNumber','CustomerId'])
```

```python
x = df.drop(columns=['Exited'])
y = df['Exited']
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_test
```

```python
x_train.info()
```

```python
x_test = x_test.drop(columns=['Surname'])
```

```python
x_train = x_train.drop(columns=['Surname'])
```

```python
X_train = scaler.fit_transform(x_train)
X_test = scaler.fit_transform(x_test)
```

```python
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train
```

```python
model = Sequential()
model.add(Dense(10,activation='sigmoid',input_dim = 11))
model.add(Dense(5,activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam')
```

```python
print(X_train.shape)
```

```python
model.fit(X_train,y_train,epochs = 15)
```

```python
model.layers[2].get_weights()
```

```python
y_log = model.predict(x_test)
y_log
```

```python
y_prd = np.where(y_log>0.5,1,0)
```

```python
y_prd
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_prd)
```

```python
from sklearn.ensemble import RandomForestClassifier
modelx = RandomForestClassifier()
```

```python
modelx.fit(x_train,y_train)
```

```python
preds = modelx.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,preds)
```
