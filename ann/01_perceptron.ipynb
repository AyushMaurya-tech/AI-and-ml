```python
from sklearn.datasets import make_classification
```

```python
import numpy as np
X,y = make_classification(n_samples=100,n_features=2,n_informative=2,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],cmap='winter',s = 100)
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis = 1)
    weights = np.ones(X.shape[1])
    lr = 0.01
    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
intercept_
```

```python
coef_
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
m
```

```python
x = np.linspace(-3,3,100)
y= m*x + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, y,color = 'red')
plt.scatter(X[:,0],X[:,1],cmap = 'winter')
plt.ylim(-3,2)
```

### loss function


```python
def gradientDescentPerceptron(X,y):
    w1 = w2 = b = 1
    lr = 0.1
    for j in range (1000):
        for i in range(X.shape[0]):
           z = w1*X[i][0] + w2*X[i][1] + b
           if z*y[i] < 0:
             w1 = w1+lr*y[i]*X[i][0]
             w2 = w2+lr*y[i]*X[i][1]
             b = b+ lr*y[i]
    return w1,w2,b
```
